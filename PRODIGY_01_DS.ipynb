```python
plt.figure(figsize = (20,20))
years = data1.columns[2:]
values = data1[years].sum()
plt.barh(years,values, color='olive',edgecolor = 'k')
plt.xlabel('Total Population',size = 15)
plt.ylabel('Years',size=15)
plt.title('Total Population per year',size = 15)
plt.show()
```
